```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# ตั้งค่าการสุ่ม
products = {
    'Electronics': ['Mouse', 'Keyboard', 'Headset', 'Monitor', 'Laptop'],
    'Clothing': ['T-shirt', 'Jeans', 'Sneakers', 'Cap'],
    'Groceries': ['Milk', 'Coffee', 'Rice', 'Water']
}
branches = ['Bangkok - Siam', 'Bangkok - Rama9', 'Chiang Mai', 'Phuket']
dates = [datetime(2024, 1, 1) + timedelta(days=x) for x in range(30)] # ข้อมูล 30 วัน

data = []
for _ in range(200): # สร้าง 200 รายการ
    category = random.choice(list(products.keys()))
    product = random.choice(products[category])
    branch = random.choice(branches)
    date = random.choice(dates)
    qty = random.randint(1, 10)
    price = random.randint(100, 20000) if category == 'Electronics' else random.randint(50, 2000)
    
    data.append([date.strftime('%Y-%m-%d'), branch, category, product, price, qty])

# สร้าง DataFrame
df = pd.DataFrame(data, columns=['Date', 'Branch', 'Category', 'Product', 'Price', 'Quantity'])
df['Total'] = df['Price'] * df['Quantity']

# Save เป็น CSV
df.to_csv('supermarket_sales.csv', index=False)
print("✅ สร้างไฟล์ 'supermarket_sales.csv' เรียบร้อย! พร้อมเข้า Power BI")
```
